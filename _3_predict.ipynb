# Predict

```python
from pathlib import Path
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

from tensorflow.keras.preprocessing import image
from tensorflow.keras.models import load_model
```

```python
# set model directory name

MODEL_DIR = Path("./model")
print(f"{MODEL_DIR= }")
```

```python
# load model

model_file = Path(MODEL_DIR, "best_model_2.12.0.h5")

model = load_model(model_file)
model
```

```python
# predict one image

cuisine_dict = {0: "salad", 1: "sushi", 2: "tofu"}

image_file = "./dataset/salad2.jpg"  # saslad(= 0)

img = image.load_img(image_file, target_size=(150, 150))
img_tensor = image.img_to_array(img)
img_tensor = np.expand_dims(img_tensor, axis=0)
img_tensor /= 255.

p = model.predict(img_tensor)[0]
ans = np.argmax(p)
print(f"{ans= }")

plt.title(cuisine_dict[ans], fontsize=12)
plt.grid()
plt.imshow(Image.open(image_file))
plt.show()
plt.close()
```

```python
# predict three images

cuisine_dict = {0: "salad", 1: "sushi", 2: "tofu"}

image_files = [
    r"./dataset/salad.jpg",  # salad(= 0)
    r"./dataset/sushi.jpg",  # sushi(= 1)
    r"./dataset/tofu.jpg",   # tofu(= 2)
]


for path in image_files:
    
    img = image.load_img(path, target_size=(150, 150))
    img_tensor = image.img_to_array(img)
    img_tensor = np.expand_dims(img_tensor, axis=0)
    img_tensor /= 255.

    p = model.predict(img_tensor)[0]
    ans = np.argmax(p)
    print(f"{ans= }")

    plt.title(cuisine_dict[ans], fontsize=12)
    plt.grid()
    plt.imshow(Image.open(path))
    plt.show()
    plt.close()
```
